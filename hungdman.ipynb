```python
def hungman(word):
#wordはプレイヤー１が当ててほしい単語をいれます。
    wrong = 0
    #プレイヤー２が間違えた回数をカウント
    
    stages = ["          ",
              "______    ",
              "|         ",
              "|    |    ",
              "|    O    ",
              "|   /|＼　",
              "|   / ＼  ",
              "|         "
             ]
    rletters = list(word)
    #wordの文字を１文字ずつの要素に分解してリストにしたもの。答えなければいけない残りの文字を覚えておくのに使用
    #ヒントの表示
    board = ["_"]*len(word)
    win = False
    print("ハングドマンへようこそ！")
    while wrong < len(stages) - 1:
    #ループを見やすくするために１行改行
        print("\n")
        char = input("１文字を予想してね（英語）")
        if char in rletters:
            cind = rletters.index(char)
            board[cind] = char
            rletters[cind] = "$"
        else:
            wrong += 1
        print(" ".join(board))
        e = wrong + 1
        print("\n".join(stages[0:e]))
        if "_" not in board:
            print("あなたの勝ちです。")
            print(" ".join(board))
            win = True
            break
    if not win:
        print("\n".join(stages[0:worng+1]))
        print("あなたの負け。正解は{}.".format(word))
    
hungman("cat")
```
